```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("meteorite_data.csv")
```

#### Q7.1 Get all the Earth meteorites that fell before the year 2000

```python
#finding all earth meteorites that fell before the year 2000
earth_meteorites_before_2000 = df[df['Year at which Earth Meteorite was hit'].dt.year < 2000]
```

#### Ans7.1 List of all earth meteorites that fell before the year 2000

```python
earth_meteorites_before_2000
```

#### Plot for the number of Earth meteorites fallen before the year 2000

```python
earth_meteorites_before_2000_count = earth_meteorites_before_2000['Year at which Earth Meteorite was hit'].dt.year.value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.plot(earth_meteorites_before_2000_count.index, earth_meteorites_before_2000_count.values)
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Number of Earth Meteorites Fallen before 2000')
plt.show()
```

#### Q7.2 Get all the earth meteorites co-ordinates who fell before the year 1970

```python
#finding all the earth meteorites co-ordinates who fell before the year 1970

filtered_df_before_1970 = df[df['Year at which Earth Meteorite was hit'].dt.year < 1970]
earth_met_coor_before_1970 = filtered_df_before_1970['point coordinates']
```

#### Ans 7.2 List of all the earth meteorites co-ordinates who fell before the year 1970

```python
earth_met_coor_before_1970
```

```python
# Scatter plot of Earth meteorites' coordinates that fell before the year 1970

# Extract x and y coordinates from the list
x_coordinates = [point[0] for point in earth_met_coor_before_1970]
y_coordinates = [point[1] for point in earth_met_coor_before_1970]


plt.figure(figsize=(5, 3))
plt.scatter(x_coordinates,y_coordinates)
plt.xlabel(x_coordinates)
plt.ylabel(y_coordinates)
plt.title('Earth Meteorites Coordinates Fallen before 1970')
plt.show()
```

#### Q7.3 Assuming that the mass of the earth meteorites was in kg, get all those whose mass was more than 10000kg

```python
earth_meteorites_mass_more_than_10000 = df[df['mass']>10000]
```

#### Ans7.3 List of all meteorites mass more than 10000

```python
earth_meteorites_mass_more_than_10000
```

```python
# Bar plot of Earth meteorites with mass more than 10000 kg
plt.figure(figsize=(10, 6))
plt.bar(earth_meteorites_mass_more_than_10000['Name of Earth Meteorite'], earth_meteorites_mass_more_than_10000['mass'])
plt.xlabel('Meteorite Name')
plt.ylabel('Mass (kg)')
plt.title('Earth Meteorites with Mass more than 10000 kg')
plt.xticks(rotation=90)
plt.show()
```
